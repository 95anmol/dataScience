```python
#cleaning data
uber_data['Request timestamp'] = pd.to_datetime(uber_data['Request timestamp'],dayfirst=True)
```

```python
#cleaning data
uber_data['Drop timestamp'] = pd.to_datetime(uber_data['Drop timestamp'],dayfirst=True)
```

```python
uber_data
uber_data.info()
```

```python
uber_data.drop_duplicates(keep="first")
```

```python
uber_data["Driver id"].fillna('-1',inplace = True)
```

```python
uber_data["request_at_hour"] = uber_data["Request timestamp"].dt.hour
```

```python
uber_data["dropped_at_hour"] = uber_data["Drop timestamp"].dt.hour
```

```python
uber_data['time_slots'] = pd.cut(uber_data['request_at_hour'], [0,5,9,12,18,21,24], labels=["Night", "Early Morning", "Late Morning", "Noon","Early Evening","Late Evening"],right=False)
```

```python
uber_data["demand"] = 1
```

```python
uber_data["supply"] = np.where(uber_data['Status'] == 'Trip Completed', 1, 0)
```

```python
uber_data["gap"] = np.where(uber_data['supply'] == 1, 0, 1)
```

```python
uber_data
```

```python
uber_data.to_csv('analysis_copy.csv')
```
