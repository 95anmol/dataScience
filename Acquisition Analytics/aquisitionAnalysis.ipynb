#### Age

```python
sns.boxplot(x='age', data=loan)
plt.show()
```

```python
# Since there are a fair few data entries with age > 71, we will club all of those ages at 71 (since we can assume 
# that the behaviour will not change significantly with age at age above 70)

loan['age'][loan['age']>71] =71
```

```python
sns.boxplot(x='age', data=loan)
plt.show()
```

```python
# To understand the behaviour of people with age, we are going to bucket the age groups into brackets of 10 and 
# study the response rates

bins = np.arange(10,loan['age'].max() + 10, 10)
loan['age_bins'] = pd.cut(loan['age'], bins, include_lowest=True)
```

```python
No_of_prospect = loan.groupby('age_bins')['response'].count().reset_index(name = 'No.of_prospect')
```

```python
count_prospects = loan[loan['response']==1].groupby('age_bins')['response'].count().reset_index(name = 'count_prospects')
```

```python
agg_age = No_of_prospect.merge(count_prospects, on='age_bins')
```

```python
agg_age['response_rate'] = 100*(agg_age['count_prospects']/agg_age['No.of_prospect']).round(3)
```

```python
agg_age
```

```python
ax = sns.barplot(x='age_bins', y='response_rate', data=agg_age)
ax.set_xticklabels(ax.get_xticklabels(),rotation=30)
plt.show()
```

As we can see, the youngest and the eldest are the most likely to buy the investment product being offered by the bank.

```python
# Note that we have added a new column named age bucket to the table

loan.head()
```

```python
# Check ROC
def draw_roc( actual, probs ):
    fpr, tpr, thresholds = metrics.roc_curve( actual, probs,
                                              drop_intermediate = False )
    auc_score = metrics.roc_auc_score( actual, probs )
    plt.figure(figsize=(5, 5))
    plt.plot( fpr, tpr, label='ROC curve (area = %0.2f)' % auc_score )
    plt.plot([0, 1], [0, 1], 'k--')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate or [1 - True Negative Rate]')
    plt.ylabel('True Positive Rate')
    plt.title('Receiver operating characteristic example')
    plt.legend(loc="lower right")
    plt.show()

    return None
```

```python
# create decile data frame
df_decile_all = pd.DataFrame(final_data)
df_decile_all.head()
```

```python
# add decile column
df_decile_all['decile'] = pd.qcut(df_decile_all['predicted_probability'], 10, labels=np.arange(10, 0, -1))
df_decile_all.head()
```

```python
# create lift  group by decile
df_lift_all = df_decile_all.groupby('decile')['predicted_probability'].count().reset_index()
df_lift_all.head()
```

```python
# rename column predicted_probability by total
df_lift_all.rename({'predicted_probability':'Total'}, axis=1, inplace=True)
df_lift_all.head()
```

```python
# add response count for each decile
df_lift_all_pred = df_decile_all[df_decile_all['actual_response']==1].groupby('decile')['actual_response'].count().reset_index()
df_lift_all_pred.head()
```

```python
# merge df_lift_all and df_lift_all_pred
df_lift_all_final = df_lift_all.merge(df_lift_all_pred,on = 'decile')
df_lift_all_final.head()
```

```python
# sort data frame
df_lift_all_final = df_lift_all_final.sort_values(['decile'], ascending=False)
df_lift_all_final.head()
```

```python
# create cummulative response
df_lift_all_final['cumresp'] = df_lift_all_final['actual_response'].cumsum()
df_lift_all_final.head()
```

```python
# create gain column
df_lift_all_final['gain'] = 100*(df_lift_all_final['cumresp']/sum(df_lift_all_final['actual_response']))
df_lift_all_final.head()
```

```python
# create cummulative lift
df_lift_all_final['cumlift'] = df_lift_all_final['gain']/(df_lift_all_final['decile'].astype('int')*(100/10))
df_lift_all_final
```
